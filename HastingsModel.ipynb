The classic Lotka-Volterra model is given to us in the form:

\begin{align*}
    & \dot{x} = x(\alpha - \beta y)\\
    & \dot{y} = y(\delta x - \gamma)
\end{align*}

The two dimensional system is restricted to the possibility of centers, spirals, and fixed points which may be stable or unstable. Run the cell below to run an animation, showing the evolution of a collection of random intial conditions in the phase space.

### Hastings Model
The Hastings model is an extension of the Lotka-Volterra model found above to three species. In addition the coeffecients are now predation rates controlled by the relative size of the populations - afterall, a predator can only hunt if there is prey to be found. 

Thus the non-dimensional Hastings model is as follows:

\begin{align*}
    & \dot{x} = x(1-x) - f_{1}(x)y\\
    & \dot{y} =  f_{1}(x)y -f_{2}(y)z - d_{1}y\\
    & \dot{z} =  f_{2}(y)z - d_{2}z\\
\end{align*}

where 

$$
f_{i}(u) = \frac{a_{i}u}{1+b_{i}u}
$$



The functions $f_{i}(u)$ are logistic growth functions. A similar function
appears in the Spruce-budworm model. If more of a prey species
is present then it becomes more likely it will be found and eaten. Thus the functions give a more
realistic model of predation in the natural world. 

Taking the authors' suggestions for parameters we have:

\begin{align*}
    & a_{1} = 5\\
    & b_{1} = 3; \\ 
    & a_{2} = 0.1\\ 
    & b_{2} = 2\\
    & d_{1} = 0.01\\
\end{align*}

The fixed points were found by Hastings and Klebanoff a few years after the inital publication.

\begin{align*}
    & (x,y,z) = (0,0,0)\\
    & (x,y,z) = (1,0,0)\\
    & (x,y,z) = \left( \frac{d_{1}}{a_{1}  - d_{1}b_{1}}, \frac{a_{1} - b_{1}d_{1}-d_{1}}{(a_{1}-b_{1}d_{1})^2} \right)\\
    & (x,y,z) = \left( 0,\frac{d_{2}}{a_{2} - b_{2}d_{2}},\frac{-d_{1}}{a_{2}-b_{2}d_{2}} \right)\\
    & (x,y,z) = \left( \hat{x_{i}} , \hat{y_{i}} , \hat{z_{i}} \right) \quad i = 1,2\\
\end{align*}

The parameters above will actually give rise to a strange attractor as we will see below. Furthermore the system also exhibits chaos, sensitivity to initial conditions.
