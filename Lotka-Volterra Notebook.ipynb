### Lotka-Volterra FoodWebs

The classic Lotka-Volterra model is given to us in the form:

\begin{align*}
    & \dot{x} = x(\alpha - \beta y)\\
    & \dot{y} = y(\delta x - \gamma)
\end{align*}

The two dimensional system is restricted to the possibility of centers, spirals, and fixed points which may be stable or unstable. 

```python
import SimpleTwoSpecies as LV
import numpy as np
%matplotlib notebook

params = [0.8, 0.1, 1.3, 0.2]
t = np.linspace(0, 15, 1000)  # time
X0 = np.array([10, 5]) 

anim = LV.LotkaVolterra(X0, params, t)
```
